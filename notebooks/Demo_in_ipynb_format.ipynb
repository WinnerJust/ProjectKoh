```python
def entropy(groups, class_values):
    #groups are rows in the left and in the right
    n_instances = float(sum([len(group) for group in groups]))
    entropy = 0.0

    #There can be no representatives of some class in one of the two groups
    #Why would be iterate over all of them
    for class_value in class_values:
        for group in groups:
            size = len(group)
            if size == 0:
              continue
            #We count proportion literally, count of some class divided by count of all
            proportion = [row[-1] for row in group].count(class_value) / size

            epsilon = 1e-10
            proportion += epsilon

            entropy += -proportion * np.log2(proportion)
    return entropy
```

```python
def _entropy(left_y, right_y):
    p_left, p_right = len(left_y), len(right_y)
    total = p_left + p_right

    unique_left, counts_left = np.unique(left_y, return_counts=True)
    unique_right, counts_right = np.unique(right_y, return_counts=True)

    prob_left = counts_left / p_left
    prob_right = counts_right / p_right


    #entropy_left = -np.sum(prob_left * np.log2(prob_left))
    #entropy_right = -np.sum(prob_right * np.log2(prob_right))
    entropy_left = -np.sum(prob_left * np.log2(prob_left + 1e-10))
    entropy_right = -np.sum(prob_right * np.log2(prob_right + 1e-10))

    entropy = (p_left / total) * entropy_left + (p_right / total) * entropy_right
    return entropy
```

```python
def get_split(x, y, feature_names, feature_history):
    b_index, b_value, b_score, b_groups = 999, 999, 999, None
    #for index in range(dataset.shape[1] - 1):

    #Trying this feature
    #We are trying every value, we don't need to trying every value
    #We have only 0 and 1
    #So python has to iterate over 10000 rows in entropy function
    #and it does it around 300 times per second
    #Which is probably good speed
    for index in range(x.shape[1]):
        #print("Trying index", index)

        feature_name_part = feature_names[index].split("_@_")[0]

        if feature_name_part in ['body_pain_location', 'pain_radiation', 'pain_characteristics', 'pain_sudden_appearance', 'pain_intensity', 'pain_localization']:
          if not np.isin('pain', feature_history):
            continue
        elif feature_name_part in ['skin_lesion_location', 'lesion_color', 'lesion_pain_intensity', 'lesion_swelling', 'lesion_itching', 'lesion_larger_than_1cm', 'lesion_peeling']:
          if not np.isin('skin_lesions', feature_history):
            continue

        for value in range(2):
            left_mask = x[:, index] == value
            left_y = y[left_mask]
            right_y = y[~left_mask]


            ent = _entropy(left_y, right_y)

            if ent < b_score:
                left_x = x[left_mask]
                right_x = x[~left_mask]
                b_index, b_value, b_score, b_left_x, b_left_y, b_right_x, b_right_y = index, value, ent, left_x, left_y, right_x, right_y


    #print("Best split is on", feature_names[b_index])
    #print("Value", b_value)
    #print("Entropy", b_score)

    #This object goes to split function and gets saved as a node using links from other nodes
    #And arguments of this function we get back from split function, that is feature_history
    return {'index': b_index, 'value': b_value, 'left_x': b_left_x, 'left_y': b_left_y, 'right_x': b_right_x, 'right_y': b_right_y, 'feature': feature_names[b_index], 'entropy': ent,
            'feature_history': np.append(feature_history, feature_names[b_index])}
```

```python
def to_terminal(group):
    #outcomes = [row[-1] for row in group]
    #return max(set(outcomes), key=outcomes.count)

    #outcomes = group[:, -1]
    unique_outcomes, counts = np.unique(group, return_counts=True)
    return unique_outcomes[np.argmax(counts)]


def split(node, feature_names, max_depth, min_size, depth):
    print("split at depth", depth, "was called")
    #left, right = node['groups']
    #del(node['groups'])
    left_x, right_x, left_y, right_y = node['left_x'], node['right_x'], node['left_y'], node['right_y']

    node['samples'] = left_x.shape[0] + right_x.shape[0]
    node['samples_check'] = left_y.shape[0] + right_y.shape[0]

    unique_values, counts = np.unique(np.concatenate((left_y, right_y)), return_counts=True)
    node['uniques'] = ""
    for value, count in zip(unique_values, counts):
      #print(f"Value: {value}, Count: {count}")
      node['uniques'] += f"{value}: {count}\n"

    del(node['left_x'])
    del(node['right_x'])
    del(node['left_y'])
    del(node['right_y'])

    #If we have all of one class in left and all of one class in right
    #But what if we have all of one class in left and lots of classes in right
    #if node['entropy'] < 1e-5:
    #    node['left'], node['right'] = to_terminal(left_y), to_terminal(right_y)
    #    return


    '''
    if not left_y.size:
        node['left'] = node['right'] = to_terminal(right_y)
        return
    if not right_y.size:
        node['left'] = node['right'] = to_terminal(left_y)
        return
    '''

    #If max depth is reached, we current node is still a splitting node
    #Be the children are class nodes already
    if depth >= max_depth:
        node['left'], node['right'] = to_terminal(left_y), to_terminal(right_y)
        return



    #Index 0 is out of bounds. WTF
    if np.all(left_y == left_y[0]):
        node['left'] = to_terminal(left_y)
        #return
    elif len(left_y) <= min_size:
        #If there are less rows than min_size
        #This is what usual tree don't have
        #But probably I just didn't set param
        node['left'] = to_terminal(left_y)
    else:
        node['left'] = get_split(left_x, left_y, feature_names, node['feature_history'])
        split(node['left'], feature_names, max_depth, min_size, depth+1)


    if np.all(right_y == right_y[0]):
        node['right'] = to_terminal(right_y)
        #return
    elif len(right_y) <= min_size:
        node['right'] = to_terminal(right_y)
    else:
        node['right'] = get_split(right_x, right_y, feature_names, node['feature_history'])
        split(node['right'], feature_names, max_depth, min_size, depth+1)

#So, tree is actually nodes that are
#{'index':b_index, 'value':b_value, 'groups':b_groups}
#Which quickly become
#{'index':b_index, 'value':b_value, 'left':left_rows, 'right':right_rows}
def build_tree(x, y, feature_names, max_depth, min_size):
    root = get_split(x, y, feature_names, np.array([]))
    split(root, feature_names, max_depth, min_size, 1)
    return root
```

```python
from graphviz import Digraph

dot = Digraph()
node_counter = 0

def generate_unique_id():
    global node_counter
    node_counter += 1
    return str(node_counter)

def draw_tree(tree, filename):
    global dot
    global node_counter

    current_id = generate_unique_id()
    dot.node(name=current_id, label=str(tree['feature']) + ' ' + str(tree['index']) + ' ' + str(tree['value']) + '\nSamples:' + str(tree['samples']) + '\nSamples check:' + str(tree['samples_check'])) # + '\n' + str(tree['uniques']))

    if isinstance(tree['left'], dict):
        #left_id = current_id + 1 #generate_unique_id()
        left_id = str(node_counter + 1)
        dot.edge(current_id, left_id)
        draw_tree(tree['left'], filename)
    else:
        left_id = generate_unique_id()
        dot.node(name=left_id, label=str(tree['left']))
        dot.edge(current_id, left_id)

    if isinstance(tree['right'], dict):
        right_id = str(node_counter + 1) #generate_unique_id()
        #Exactly this ID will be generated in draw_tree call
        dot.edge(current_id, right_id)
        draw_tree(tree['right'], filename)
    else:
        #Here we generate right here, because there is no draw_tree call which will generate ID
        right_id = generate_unique_id()
        dot.node(name=right_id, label=str(tree['right']))
        dot.edge(current_id, right_id)
```

```python
import numpy as np

class DecisionTree:
    def __init__(self, max_depth=None, min_samples_split=2, min_samples_leaf=1):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        self.tree = {}

    def fit(self, X, y, depth=0):
        '''
        if len(set(y)) == 1:
            return {'class': y[0]}

        if self.max_depth is not None and depth >= self.max_depth:
            return {'class': max(set(y), key=y.count)}

        if len(X) == 0:
            return {'class': None}
        '''
        if len(np.unique(y)) == 1 or len(X) < self.min_samples_split or \
                (self.max_depth is not None and depth >= self.max_depth):
            return {'class': np.argmax(np.bincount(y))}

        if len(X) < self.min_samples_leaf:
            return {'class': np.argmax(np.bincount(y))}


        # Find the best split
        best_index, best_value = self._find_best_split(X, y)


        #if split_index is None:
        #    return {'class': max(set(y), key=y.count)}

        if best_index is None:
            return {'class': np.argmax(np.bincount(y))}

        #left_indices = [i for i, val in enumerate(X[:, split_index]) if val <= split_value]
        #right_indices = [i for i, val in enumerate(X[:, split_index]) if val > split_value]
        left_indices = X[:, best_index] <= best_value
        right_indices = X[:, best_index] > best_value

        left_tree = self.fit(X[left_indices], y[left_indices], depth + 1)
        right_tree = self.fit(X[right_indices], y[right_indices], depth + 1)

        return {'index': best_index,
                'value': best_value,
                'left': left_tree,
                'right': right_tree}

    def _find_best_split(self, X, y):
        best_index, best_value, best_entropy = None, None, float('inf')

        for index in range(X.shape[1]):
            sorted_indices = np.argsort(X[:, index])
            unique_values = np.unique(X[sorted_indices, index])

            '''
            for value in set(X[:, index]):
                left_indices = [i for i, val in enumerate(X[:, index]) if val <= value]
                right_indices = [i for i, val in enumerate(X[:, index]) if val > value]

                if len(left_indices) == 0 or len(right_indices) == 0:
                    continue
                    '''
            for value in unique_values:
                left_indices = X[:, index] <= value
                right_indices = X[:, index] > value

                if len(np.unique(y[left_indices])) < self.min_samples_leaf or len(np.unique(y[right_indices])) < self.min_samples_leaf:
                    continue

                entropy = self._entropy(y[left_indices], y[right_indices])
                if entropy < best_entropy:
                    best_index, best_value, best_entropy = index, value, entropy

        return best_index, best_value

    def _entropy(self, left_y, right_y):
        p_left, p_right = len(left_y), len(right_y)
        total = p_left + p_right

        #prob_left = np.array([(left_y.count(i) / p_left) for i in set(left_y)])
        #prob_right = np.array([(right_y.count(i) / p_right) for i in set(right_y)])
        prob_left = np.array([(left_y == c).sum() / p_left for c in np.unique(left_y)])
        prob_right = np.array([(right_y == c).sum() / p_right for c in np.unique(right_y)])


        #entropy_left = -np.sum(prob_left * np.log2(prob_left))
        #entropy_right = -np.sum(prob_right * np.log2(prob_right))
        entropy_left = -np.sum(prob_left * np.log2(prob_left + 1e-10))
        entropy_right = -np.sum(prob_right * np.log2(prob_right + 1e-10))

        entropy = (p_left / total) * entropy_left + (p_right / total) * entropy_right
        return entropy

    def predict(self, X):
        predictions = []
        for sample in X:
            prediction = self._predict_tree(sample, self.tree)
            predictions.append(prediction)
        return predictions

    '''
    def _predict_tree(self, sample, tree):
        if 'class' in tree:
            return tree['class']

        index, value = tree['index'], tree['value']
        if sample[index] <= value:
            return self._predict_tree(sample, tree['left'])
        else:
            return self._predict_tree(sample, tree['right'])
            '''

    def _predict_tree(self, sample, tree):
        if 'class' in tree:
            return tree['class']

        if 'index' not in tree or 'value' not in tree:
            # If necessary keys are absent, return the majority class
            return np.argmax(np.bincount(tree['class']))

        index, value = tree['index'], tree['value']
        if sample[index] <= value:
            return self._predict_tree(sample, tree['left'])
        else:
            return self._predict_tree(sample, tree['right'])
```

```python
def dot_to_json(file_in, file_out, indent=1):
    import json
    try:
        import networkx
        import pygraphviz
        from networkx.readwrite import json_graph
    except ImportError:
        print("Install pygraphviz and networkx:")
        print("sudo apt-get install python-pygraphviz python-networkx")
        return 1

    graph_dot  = pygraphviz.AGraph( file_in )
    graph_netx = networkx.from_agraph( graph_dot )
    graph_json = json_graph.node_link_data( graph_netx )

    # fix formatting [graphviz to d3.js]
    for node in graph_json["nodes"]:
        # replace label by name
        node['name'] = node.pop('label')
        # id from string to integer
        node['id']   = int(node['id'])

    with open(file_out, 'w') as f:
        json.dump(graph_json, f, indent=indent)

    return 0

dot_to_json('graph.dot', 'graph.json')
```

```python
def treeToJson(decision_tree, feature_names=None):
  from warnings import warn

  js = ""

  def node_to_str(tree, node_id, criterion):
    if not isinstance(criterion, sklearn.tree.tree.six.string_types):
      criterion = "impurity"

    value = tree.value[node_id]
    if tree.n_outputs == 1:
      value = value[0, :]

    jsonValue = ', '.join([str(x) for x in value])

    if tree.children_left[node_id] == sklearn.tree._tree.TREE_LEAF:
      return '"id": "%s", "criterion": "%s", "impurity": "%s", "samples": "%s", "value": [%s]' \
              % (node_id,
                criterion,
                tree.impurity[node_id],
                tree.n_node_samples[node_id],
                jsonValue)
    else:
      if feature_names is not None:
        feature = feature_names[tree.feature[node_id]]
      else:
        feature = tree.feature[node_id]

      if "=" in feature:
        ruleType = "="
        ruleValue = "false"
      else:
        ruleType = "<="
        ruleValue = "%.4f" % tree.threshold[node_id]

      return '"id": "%s", "rule": "%s %s %s", "%s": "%s", "samples": "%s"' \
              % (node_id,
                feature,
                ruleType,
                ruleValue,
                criterion,
                tree.impurity[node_id],
                tree.n_node_samples[node_id])

  def recurse(tree, node_id, criterion, parent=None, depth=0):
    tabs = "  " * depth
    js = ""

    left_child = tree.children_left[node_id]
    right_child = tree.children_right[node_id]

    js = js + "\n" + \
          tabs + "{\n" + \
          tabs + "  " + node_to_str(tree, node_id, criterion)

    if left_child != sklearn.tree._tree.TREE_LEAF:
      js = js + ",\n" + \
            tabs + '  "left": ' + \
            recurse(tree, \
                    left_child, \
                    criterion=criterion, \
                    parent=node_id, \
                    depth=depth + 1) + ",\n" + \
            tabs + '  "right": ' + \
            recurse(tree, \
                    right_child, \
                    criterion=criterion, \
                    parent=node_id,
                    depth=depth + 1)

    js = js + tabs + "\n" + \
          tabs + "}"

    return js

  if isinstance(decision_tree, sklearn.tree.BaseDecisionTree):
    js = js + recurse(decision_tree, 0, criterion="impurity")
  else:
    js = js + recurse(decision_tree.tree_, 0, criterion=decision_tree.criterion)

  return js
```

```python
import numpy as np

from sklearn.tree import _tree


def export_json(decision_tree, out_file=None, feature_names=None):
    """Export a decision tree in JSON format.
    This function generates a JSON representation of the decision tree,
    which is then written into `out_file`. Once exported, graphical renderings
    can be generated using, for example::
        $ dot -Tps tree.dot -o tree.ps      (PostScript format)
        $ dot -Tpng tree.dot -o tree.png    (PNG format)
    Parameters
    ----------
    decision_tree : decision tree classifier
        The decision tree to be exported to JSON.
    out : file object or string, optional (default=None)
        Handle or name of the output file.
    feature_names : list of strings, optional (default=None)
        Names of each of the features.
    Returns
    -------
    out_file : file object
        The file object to which the tree was exported.  The user is
        expected to `close()` this object when done with it.
    Examples
    --------
    >>> from sklearn.datasets import load_iris
    >>> from sklearn import tree
    >>> clf = tree.DecisionTreeClassifier()
    >>> iris = load_iris()
    >>> clf = clf.fit(iris.data, iris.target)
    >>> import tempfile
    >>> out_file = tree.export_json(clf, out_file=tempfile.TemporaryFile())
    >>> out_file.close()
    """
    import numpy as np

    from sklearn.tree import _tree

    def arr_to_py(arr):
        arr = arr.ravel()
        wrapper = float
        if np.issubdtype(arr.dtype, np.int):
            wrapper = int
        return map(wrapper, arr.tolist())


    def node_to_str(tree, node_id):
        node_repr = '"error": %.4f, "samples": %d, "value": %s' \
                    % (tree.init_error[node_id],
                       tree.n_samples[node_id],
                       arr_to_py(tree.value[node_id]))
        if tree.children_left[node_id] != _tree.TREE_LEAF:
            if feature_names is not None:
                feature = feature_names[tree.feature[node_id]]
            else:
                feature = "X[%s]" % tree.feature[node_id]

            label = '"label": "%s <= %.2f"' % (feature,
                                               tree.threshold[node_id])
            node_type = '"type": "split"'
        else:
            node_type = '"type": "leaf"'
            label = '"label": "Leaf - %d"' % node_id
        node_repr = ", ".join((node_repr, label, node_type))
        return node_repr

    def recurse(tree, node_id, parent=None):
        if node_id == _tree.TREE_LEAF:
            raise ValueError("Invalid node_id %s" % _tree.TREE_LEAF)

        left_child = tree.children_left[node_id]
        right_child = tree.children_right[node_id]

        # Open node with description
        out_file.write('{%s' % node_to_str(tree, node_id))

        # write children
        if left_child != _tree.TREE_LEAF:  # and right_child != _tree.TREE_LEAF
            out_file.write(', "children": [')
            recurse(tree, left_child, node_id)
            out_file.write(', ')
            recurse(tree, right_child, node_id)
            out_file.write(']')

        # close node
        out_file.write('}')

    if out_file is None:
        out_file = open("tree.json", "w")
    elif isinstance(out_file, basestring):
        out_file = open(out_file, "w")

    if isinstance(decision_tree, _tree.Tree):
        recurse(decision_tree, 0)
    else:
        recurse(decision_tree.tree_, 0)

    return out_file
```
