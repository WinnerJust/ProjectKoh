```python
def convert_to_title_case(name):
    words = name.split('_')
    words[0] = words[0].capitalize()
    return ' '.join(words)

print(convert_to_title_case("nodal_skin_eruptions"))
print(convert_to_title_case("itching"))
```

```python
import re

def find_questions(text):
    # Define a regular expression pattern to find questions
    pattern = r'(\b[A-Za-z][^\?]*\?\b)'

    # Use re.findall to extract all the questions from the text
    questions = re.findall(pattern, text)

    return questions
```

```python
import time
import json
```

```python
questions = {}
```

```python
for symptom in df.columns:
  if any(char.isdigit() for char in symptom):
    continue

  while symptom not in questions or len(questions[symptom]) < 50:

    if symptom in questions:
      print("Amount of questions for", symptom, ":", len(questions[symptom]))
    else:
      print("Amount of questions for", symptom, ": 0")

    #print(symptom)

    title_symptom_name = convert_to_title_case(symptom)
    #print(title_symptom_name)

    resp = gpt(f"Do you have {title_symptom_name}? Paraphrize this question. Provide 100 paraphrized questions in a numbered list").replace('"', '')

    print(resp)

    #parsed = find_questions(resp)

    parsed = re.findall(r'\d+\.\s(.*?\?)', resp)
    print("parsed:")
    print(parsed)

    if symptom not in questions:
      questions[symptom] = parsed
    else:
      questions[symptom].extend(parsed)


    file_name = "questions.txt"
    with open(file_name, 'w') as file:
        file.write(json.dumps(questions, indent=4))

    time.sleep(60)
```
