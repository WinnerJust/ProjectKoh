```python
file_path = '35829017'
with open(file_path, 'r') as file:
    json_data = json.load(file)

print(json_data)
```

```python
import re

#Translating values for evidences
for cond in json_data:
  for value in json_data[cond]["value_meaning"]:
    eng_name = json_data[cond]["value_meaning"][value]["en"]
    new_value = re.sub(r'[^a-zA-Z0-9]', '_', eng_name)

    #print(eng_name, new_value)

    if value in map:
      print("Already exists:", value)

    map[value] = new_value
```

```python
#Making json for a prompt to ChatGPT to translate short names for evidences from French to English using translationg of questions that was given with the dataset
data_for_gpt = []

for cond in json_data:
  data_for_gpt.append({
      "shortname_fr": cond,
      "question_fr": json_data[cond]["question_fr"],
      "question_en": json_data[cond]["question_en"],
      "shortname_en": "",
      "is_antecedent": json_data[cond]["is_antecedent"]
  })

data_for_gpt
```

```python
#Check of uniqueness in output from gpt
#We should not have the same translation to english for some french words

keys = []
values = []

with open("french_english_map.txt", 'r') as file:
    for line in file:
        key, value = line.strip().split(': ')

        keys.append(key)
        values.append(value)


def are_all_unique(arr):
    unique_values = set()

    for value in arr:
        if value in unique_values:
            print("Impostor:", value)
            return False
        else:
            unique_values.add(value)

    return True

print(are_all_unique(keys))
print(are_all_unique(values))
```

```python
with open("french_english_map.txt", 'r') as file:
    for line in file:
        key, value = line.strip().split(': ')

        for entry in data_for_gpt:
          if entry["shortname_fr"] == key:
            new_value = re.sub(r'[^a-zA-Z0-9_]', '', value)

            if entry["is_antecedent"]:
              new_value = "antecedent_" + new_value
            #if new_value != value:
            #  print("Changed", value, "to", new_value)
            entry["shortname_en"] = new_value
```

```python
#Check of uniqueness after anticedent keyword has been added
#Previous check is useless then

keys = []
values = []

for entry in data_for_gpt:
  keys.append(entry["shortname_fr"])
  values.append(entry["shortname_en"])

print(are_all_unique(keys))
print(are_all_unique(values))
```

```python
#Checking that we don't have shortnames that were not translated
filtered_array = [item for item in data_for_gpt if 'shortname_en' in item and item['shortname_en'] == '']

filtered_array
```

```python
import json

df['EVIDENCES'] = df['EVIDENCES'].apply(lambda x: json.loads(x.replace("'", '"')))

df.head()
```

```python
#Performing translation

for index, row in df.iterrows():
  if row['PATHOLOGY'] not in map:
    print("No translation for:", row['PATHOLOGY'])
  else:
    #row['PATHOLOGY'] = map[row['PATHOLOGY']]
    df.at[index, 'PATHOLOGY'] = map[row['PATHOLOGY']]

  for i, evidence in enumerate(row['EVIDENCES']):
    evidence_arr = row['EVIDENCES'][i].split("_@_")

    new_evidence_arr = []

    for j, evidence_part in enumerate(evidence_arr):
      if evidence_part not in map:
        print("No translation for:", evidence_part)
      else:
        new_evidence_arr.append(map[evidence_part])

    #print(row['EVIDENCES'][i], "_@_".join(new_evidence_arr))
    row['EVIDENCES'][i] = "_@_".join(new_evidence_arr)
```

```python
#Some addition information about the dataset

from collections import Counter

pathology_evidence_counts = {}

#Iterate through the DataFrame and count evidences for each pathology
for index, row in df.iterrows():
    pathology = row['PATHOLOGY']
    evidences = row['EVIDENCES']
    evidence_count = Counter(evidences)

    if pathology not in pathology_evidence_counts:
        pathology_evidence_counts[pathology] = {}

    for evidence, count in evidence_count.items():
        if evidence in pathology_evidence_counts[pathology]:
            pathology_evidence_counts[pathology][evidence] += count
        else:
            pathology_evidence_counts[pathology][evidence] = count

#Display the mapping of pathology to evidence counts
counter = 1
for pathology, evidence_count in pathology_evidence_counts.items():
    print(f"Pathology {counter}: {pathology}")
    counter += 1

    #sorted_evidences = sorted(evidence_count.items(), key=lambda x: x[0])  # Sort by evidence name

    for evidence, count in evidence_count.items():
        print(f" - {evidence}: {count}")
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
```

```python

x_train, x_test, y_train, y_test = train_test_split(data_feature, data_label, test_size=0.2, random_state=42)

x_train = x_train.reset_index(drop=True)
x_test = x_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

gc.collect()
```

```python
clf = None
gc.collect()

clf = DecisionTreeClassifier(criterion='entropy', max_depth=15) #To make it as it was before #, min_samples_split=5)

clf.fit(x_train, y_train)
#clf.fit(data_feature, data_label)

y_pred = clf.predict(x_test)
```

```python
clf = None
gc.collect()

#Trying parameters we got from GridSearch
params = {
    'criterion': 'entropy',
    'max_depth': 15,
    'max_features': None, #'sqrt',
    'max_leaf_nodes': None,
    'min_impurity_decrease': 0.0,
    'min_samples_leaf': 1,
    'min_samples_split': 5,
    'random_state': 42,
    'splitter': 'best'
}

clf = DecisionTreeClassifier(**params)

clf.fit(x_train, y_train)
#clf.fit(data_feature, data_label)

y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    #'max_depth': [3, 5, 7, 10],
    'max_depth': [10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2'],
    'max_leaf_nodes': [None, 5, 10, 20],
    'min_impurity_decrease': [0.0, 0.1, 0.2],
    'random_state': [42],
    'splitter': ['best', 'random']
}

# Create a decision tree classifier
dt_classifier = DecisionTreeClassifier()

# Use GridSearchCV to search for the best parameters
grid_search = GridSearchCV(estimator=dt_classifier, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2)
grid_search.fit(x_train, y_train)

# Get the best parameters and the best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Score:", best_score)
```
